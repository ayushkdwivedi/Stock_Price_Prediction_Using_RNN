```python
import math
from decimal import *
getcontext().prec = 7
```

```python
sequence =   "CTTCATGTGAAAGCAGACGTAAGTCA"
print("Input sequence is:")
print(sequence)
print("The sequence as a list")
list_sequence = list(sequence)
print(list_sequence)
state_path = ("EEEEEEEEEEEEEEEEEE5IIIIIII$")
print("State path is")
print (state_path)
print("The state path as a list")
list_state_path = list(state_path)
print(list_state_path)
list_of_nucleotides = ["A", "C", "G", "T"]
```

```python
dict_prob_emission = {"E":
                      { "A":0.25,
                        "C":0.25,
                        "G":0.25,
                       "T":0.25 }
                     }
dict_prob_emission.update({"5":
                      { "A":0.05, 
                       "C":0.0, 
                       "G":0.95, 
                       "T":0.0}}) 
dict_prob_emission.update({"I":
                      {"A":0.4, 
                       "C":0.1, 
                       "G":0.1, 
                       "T":0.4}})
print(" the nucleotides are")
print(list_of_nucleotides)
print("Probability emission rate are")
print(dict_prob_emission)
dict_prob_transition = { "^" : { "E" : 1 } }       
dict_prob_transition.update({"E" : { "E" : 0.9 , "5" : 0.1 } })  
dict_prob_transition.update({ "5" :{ "I" : 1 } })                        ,
dict_prob_transition.update({ "I" :{ "I" : 0.9 , "$" : 0.1 }})
print("Probability transmission rate are")
print(dict_prob_transition)

```

#### What we are applying is known as the forward backward theorem. 
#### the algorithm involves three steps:
#### 1. Computing forward probabilities
#### 2. Computing backward probabilities
#### 3. Computing smoothed values.

```python
prob = dict_prob_transition["^"][state_path[0]]
print(prob)

for i in range(0, len(list_sequence)):
    n = list_sequence[i]
    s0 = state_path[i]
    e = dict_prob_emission[s0][n]
    s1 = state_path[i+1]
    t = dict_prob_transition[s0][s1]
    print("State Path, sequence, probability emission")
    print(" s0:",s0," n:",n," e:",e)
    print("State Path, Next State Path, transition probability")
    print(" s0:",s0," s1:",s1," t:",t)

    prob = Decimal(prob)
    prob*= Decimal(e) 
    prob*= Decimal(t)
    print(prob)
    print(math.log(prob))
print("FINAL PROBABILITY")
print(math.log(prob))
```
